```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('../data/data_from_uefa.xlsx', encoding = 'iso-8859-1')
```

```python
df = df.drop(columns = ['URL']).rename(columns = {'Team_Home' : 'team_home',
                                                  'Team_Away' : 'team_away',
                                                  'Date' : 'date'
                                                 })
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
cols_to_transform = ['possession_home', 'possession_away']

for col in cols_to_transform:
    df[col] = df[col].apply(lambda possession_string : possession_string[:-1]).astype(int)
```

```python
cols_teams = ['team_home', 'team_away']

for col in cols_teams:
    df[col] = df[col].apply(lambda team : team if team != 'FYR Macedonia' else 'North Macedonia')
```

```python
def change_name_home_away(name):
    if name[-4:] == 'home':
        return name[:-4] + 'away'
    elif name[-4:] == 'away':
        return name[:-4] + 'home'
    else:
        return name
```

```python
cols_mirrored = [change_name_home_away(col) for col in df.columns]
```

```python
df_mirrored = df.copy()
df_mirrored.columns = cols_mirrored
```

```python
df_train = pd.concat([df, df_mirrored]).reset_index(drop = True)
```

```python
cols_to_keep = [col for col in df_train.columns if '_away' not in col] 
```

```python
cols_to_keep_2 = [col[:-5] if col[-5:] == '_home' else col for col in cols_to_keep]
```

```python
def weighted_mean(c, arr):
    n = len(arr)
    
    weights = [2**(-c*m) for m in range(0, n)][::-1]
    
#     print(weights[1:])
#     print(arr[:-1])
    
    return np.dot(weights[1:], arr[:-1])/sum(weights)
```

```python
def normal_mean(arr):
    n = len(arr)
    
    return sum(arr[:-1])/(n-1)
```

```python
def xxxyyy(df_train, country):
    df = df_train.copy()
    df = df[df['team'] == country].sort_values(by = ['date'], ascending = [1]).reset_index(drop = True)
    
    df_1 = pd.concat([df.iloc[:1] for i in range(10)])
    df_2 = df.iloc[1:]
    
    df_x = pd.concat([df_1, df_2]).reset_index(drop = True)
    
    col_feat = [col for col in df_train.columns if col not in ['team', 'date']]
    
    for col in col_feat:
        df_x[col + '_weighted_mean_5']  = df_x[col].rolling(5+1).apply(lambda arr : weighted_mean(0.25, arr))
        df_x[col + '_weighted_mean_10'] = df_x[col].rolling(10+1).apply(lambda arr : weighted_mean(0.25, arr))
        
        df_x[col + '_normal_mean_5']  = df_x[col].rolling(5+1).apply(lambda arr : normal_mean(arr))
        df_x[col + '_normal_mean_10'] = df_x[col].rolling(10+1).apply(lambda arr : normal_mean(arr))
    
    df_x = df_x[10:].reset_index(drop = True)
    
    return df_x
```

```python
dfuu = xxxyyy(df_train, 'Albania')
```
